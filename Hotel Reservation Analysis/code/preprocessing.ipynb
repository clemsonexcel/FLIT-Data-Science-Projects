```python
# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pymysql
from sqlalchemy import create_engine
```

**Null Values**

* There are a total of four columns with missing values; children, country, agent and comapany.
* children column has 4 missing valies
* country column has 488 missing
* agent column has 16340 missing values,
* comapany has 112593 missing values

**Filling the null values**

```python
# Fill null values in 'country' with unknown
data['country'] = data['country'].fillna('unknown')

# Fill null values in 'agent' with 0
data['agent'] = data['agent'].fillna(0,)


# Fill null values in 'company' with 0
data['company'] = data['company'].fillna(0)


# Fill null values in 'children' with 0
data['children'] = data['children'].fillna(0)
```
